```python
import numpy as np
```

```python
class Tensor(np.ndarray):
    def __new__(cls, input_array, op_name=None) -> Tensor:
        cls.name = op_name
        return np.asarray(input_array, dtype=tensor_type()).view(cls)

    def __array_finalize__(self, obj) -> None:
        if obj is None:
            return
```

```python
from absl import flags
import sys

# Check if the flag is already defined before defining it
if 'tensor_width' not in flags.FLAGS:
    flags.DEFINE_integer('tensor_width', 64, 'Width of complex (64,128)')

# This line is often needed in notebooks to "finish" the setup
# so you can actually read the values.
flags.FLAGS(sys.argv, known_only=True)

def tensor_width():
    return flags.FLAGS.tensor_width

def tensor_type():
    assert tensor_width() == 64 or tensor_width() == 128
    return np.complex64 if tensor_width() == 64 else np.complex128
```
